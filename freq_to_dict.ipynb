```python
import json 

with open("frequency.json", "r") as file:
    words = json.load(file)
words
```

```python
words[]
```

```python
token_id_pairs = {}
i = 1
for word in words:
    if i < 20001:
        token_id_pairs[word] = i
        i += 1
token_id_pairs
```

```python
with open("tokenizer_dict_v1.json", "w") as f:
    json.dump(token_id_pairs, f, ensure_ascii=False)
    f.close
```
